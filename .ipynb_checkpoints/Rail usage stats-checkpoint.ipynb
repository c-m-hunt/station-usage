# Estimates of station usage

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
pd.options.display.float_format = '{:,.2f}'.format
```

```python
total_col = '1718 Entries & Exits'
last_year_total_col ='1617 Entries & Exits'
change_col = 'Year on Year Change'
season_col = 'Season Percent'
df = pd.read_excel('data/estimates-of-station-usage-2017-18.xlsx', sheet_name='Estimates of Station Usage')
```

### Tidy and preview the data

```python
# Convert columns to numbers
num_cols = [total_col, last_year_total_col, 'Entries & Exits_Season', 'Entries & Exits_Reduced', 'Entries & Exits_Full']

for col in num_cols:
    df[col] = pd.to_numeric(df[col], errors='coerce')
    
df[change_col] = ((df[total_col] - df[last_year_total_col]) / df[last_year_total_col])
df[season_col] = df['Entries & Exits_Season'] / df[total_col]
df.head()
```

## Stations

### Top stations by entries and exits

```python
top_stations=df.sort_values([total_col], ascending=False).head(20)[['Station Name','Region', total_col]]
top_stations
```

```python
fig, ax = plt.subplots(figsize=(15,7))
ax = sns.barplot(data=top_stations, x='Station Name', y=total_col, ax=ax, palette='rocket')
ax = ax.set_xticklabels(labels=top_stations['Station Name'],rotation=45, ha='right')
```

### And the smallest

```python
df.sort_values([total_col], ascending=True).head(5)[['Station Name','Region', total_col]]
```

### Distribution of station entry and exits
Most stations have a very small turnover (Network Rail stations excluded)

```python
fig, ax = plt.subplots(figsize=(15,5))
ax = sns.distplot(df[df['Station Facility Owner'] != 'Network Rail'][total_col], hist=True, ax=ax, bins=100, kde=False)
```

### Largest growing stations

```python
df[['Station Name', 'Region', 'Station Facility Owner', total_col, last_year_total_col, change_col]] \
    .sort_values(change_col, ascending=False).head(10)
```

### Most declining stations

```python
df[['Station Name', 'Region', 'Station Facility Owner', total_col, last_year_total_col, change_col]] \
    .sort_values(change_col, ascending=True).head(10)
```

## Station owners

```python
owners = df.groupby(['Station Facility Owner'])[total_col].agg(['sum', 'count', 'mean']).sort_values(['sum'], ascending=False)
owners.rename(index=str, columns={'sum': 'Total', 'count': 'Stations', 'mean': 'Ave'}, inplace=True)
owners.reset_index(inplace=True)
owners.index += 1
owners
```

## Routes

### Commuters (Season tickets)

```python
routes = df.groupby(['SRS Description'])['Entries & Exits_Season', total_col].agg(['sum', 'count', 'mean'])
routes['Commuter Percent'] = (routes['Entries & Exits_Season']['sum'] / routes[total_col]['sum'] * 100)
routes.sort_values(['Commuter Percent'], ascending=False).head(10)
```

### Route growth

```python
routes = df.groupby(['SRS Description'])[last_year_total_col, total_col].agg(['sum', 'max', 'count', 'mean'])
routes['Growth percent'] = (routes[last_year_total_col]['sum'] / routes[total_col]['sum'] * 100) - 100
routes.sort_values(['Growth percent'], ascending=False).head(20)
```

### Route decline

```python
routes.sort_values(['Growth percent'], ascending=True).head(20)
```
